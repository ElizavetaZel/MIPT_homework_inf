```python
def cities(city):
    first = []
    last = []
    q = 0
    for i in range(len(city)):
        city_new = list(city[i])
        first.append(city_new[0])
        last.append(city_new[len(city_new) - 1])
    for i in first:
        for j in last:
            if i == j:
                q += 1
                j = "*"
                break
    if q == len(city) - 1:
        return True
    else:
        return False
            

cities1 = ["london", "newyork", "klin", "vienna"] 
cities2 = ["paris", "sydney", "york", "kyoto"]  
cities3 = ["paris", "london", "moscow"]
print(cities(cities1))
print(cities(cities2))
print(cities(cities3))
```
